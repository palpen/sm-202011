```python
train['age'] = train['age'].apply(lambda x : round(x))
```

```python
category = pd.cut(train.age,bins=[0,20,39,59,79,99],labels=['<20','20-39','40-59','60-79','80+'])
train.insert(6,'Age Group',category)
```

```python
category = pd.cut(train.bmi,bins=[0,18.5,24.9,29.9,97.6],labels=['Under weight','Normal weight','Over weight','Obesity'])
train.insert(6,'BMI',category)
```

```python
train = train.dropna()
print(train)
```
